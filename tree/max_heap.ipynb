# 最大堆

优先队列（Priority Queue）：基于元素优先级大小来取元素，而非传统队列的FIFO模式。

基于完全二叉树来实现。

最大堆的有序特性：每个节点都大于其左、右子树的所有元素。


```python
class maxHeap(object):
    def __init__(self, maxSize):
        self.data = []
        self.size = 0
        self.cap = maxSize
        ## 哨兵，将其定义为大于堆中所有可能元素的值
        self.data.append(32767)
    
    def isFull(self):
        return self.size == self.cap
    
    def isEmpty(self):
        return self.size == 0
    
    def insert(self, x):
        if self.isFull():
            return False
        self.size += 1
        self.data.append(x)
        i = self.size
        while self.data[i//2] < x:
            self.data[i] = self.data[i//2]
            i = i // 2
        self.data[i] = x
        return True
            
    def deleteMax(self):
        """
        时间复杂度：O(log(n))
        """
        if self.isEmpty():
            return False
        self.size -= 1
        maxElem = self.data[1]
        x = self.data.pop(-1)
        i = 1
        while 2*i < self.size:
            tmp = self.data[2*i]
            j = 2*i
            if (2*i + 1) <= self.size and self.data[2*i] < self.data[2*i+1]:
                tmp = self.data[2*i+1]
                j = 2*i+1
            if tmp > x:
                self.data[i] = tmp
                i = j
                self.data[j] = x
            else:
                self.data[i] = x
                break
        #self.data[i] = x
        return maxElem
    
    def build(self, data_list):
        def perc_down(i):
            x = self.data[i]
            while 2*i <= self.size:
                tmp = self.data[2*i]
                j = 2*i
                if (2*i + 1) <= self.size and self.data[2*i] < self.data[2*i+1]:
                    tmp = self.data[2*i+1]
                    j = 2*i+1
                if tmp > x:
                    self.data[i] = tmp
                    i = j
                    self.data[j] = x
                else:
                    self.data[i] = x
                    break                
        if len(data_list) > self.cap:
            raise Exception("the length to be inserted is larger than capacity")
        self.data.extend(data_list)
        self.size += len(data_list)
        i = self.size // 2
        while i >= 1:
            perc_down(i)
            i = i - 1
        

# ha = maxHeap(10)
# ha.insert(10)
# ha.insert(8)
# ha.insert(9)
# ha.insert(4)
# ha.insert(5)
# ha.insert(11)
# print(ha.data)
# print(ha.deleteMax())
# print(ha.data)
hh = maxHeap(18)
x = list(range(10))
print(x)
hh.build(x)
print(hh.data)
print(hh.size)
```
